```python
# 4.4.7.complement.py 
seq = "TATAAAGGCAATATGCAGTAG" 
comp_dic = { 'A':'T', 'C':'G', 'G':'C', 'T':'A' }  # 상보적 염기를 키-값 으로 하는 사전을 만든다 
comp_seq = "" 
for s in seq:  # 서열에서 하나씩 읽어서 
    comp_seq += comp_dic[s]  # 상보적 염기를 추가해준다 
revcomp_seq = comp_seq[::-1]  # 파이썬 문자열을 뒤집어준다 
print(comp_seq)  # ATATTTCCGTTATACGTCATC 
print(revcomp_seq)  # CTACTGCATATTGCCTTTATA 
```

```python
seq = Seq("TATAAAGGCAATATGCAGTAG") 
comp_seq = seq.complement() # =template strand DNA
rev_comp_seq = seq.reverse_complement() 
print(comp_seq)  # ATATTTCCGTTATACGTCATC 
print(rev_comp_seq)  # CTACTGCATATTGCCTTTATA 
```

```python
# 4.4.9.orf_finder.py

# tatabox_seq = Seq("tataaaggcAATATGCAGTAG")
start_idx = tatabox_seq.find("ATG") # find 메서드는 index number를 리턴
print(start_idx)
end_idx = tatabox_seq.find("TAG", start_idx) # 예문의 편의상 TAG 로, find시작을 start_idx부터
print(end_idx)
orf = tatabox_seq[start_idx:end_idx+3] # 파이썬 문자열과 같은 방법으로 슬라이싱이 가능하다. 
print(orf)  # ATGCAGTAG
```

```python
# 4.5.2.calc_molecular_weight.py 
from Bio.Seq import Seq
from Bio.SeqUtils import molecular_weight 

seq1 = Seq("ATGCAGTAG") 

print(molecular_weight(seq1))  # default = DNA, 2842.82 출력 
print(molecular_weight(seq1, "DNA")) # 명시적으로 seq_type="DNA" 로 명시
print(molecular_weight(seq1.transcribe(), seq_type="RNA")) # seq_type="RNA"  
print(molecular_weight(seq1, "protein")) # seq_type="protein"
```
